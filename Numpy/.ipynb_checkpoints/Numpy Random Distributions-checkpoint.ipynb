# Normal Distributions

* Use random.normal() to get Normal data Distribution.


It has 3 parameters.

1. loc-(Mean)where peak of the data exist.
2. scale- (Standard Deviation) how flat the graph should be
3. size- Shape of returned array.

```python
import numpy as np
nd=np.random.normal(loc=5,scale=0.01,size=16)
print(nd)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.distplot(np.random.normal(size=1000,loc=5,scale=0.2),hist=True)
```

# Binomial Distribution

It is a discrete distribution for which every event will have two possible outcomes(binary outcomes).Eg:tossing a coin,pass/fail in test.
Parameters:
* We can use random.binomial() to get binomial data distributon.

1. n - no. of trials
2. p- probability of occurence of each outcome
3. N - size of returned array

```python
import numpy as np
nd=np.random.binomial(n=10,p=0.5,size=200)
print(nd)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.distplot(np.random.normal(size=1000,loc=50,scale=5),hist=True)
sns.distplot(np.random.binomial(size=1000,n=100,p=0.5),hist=True)
plt.show()
```

How is Normla Distribution different from Binomail Distribution?

Normal Distribution is for Continuous data
Binomial for Discrete Data
But if there enough data points it approximates binomial distribution with certain loc and scale.
