```python
import pandas as pd
df=pd.DataFrame({"Students":["Ram","Suresh","Ramesh","Joe","Monica"],"Marks":[90,47,89,24,98]})
df["Results"]=df.Marks.apply(lambda x: "Pass"if x>50 else "Fail")
```

```python
pd.crosstab(df.Students,df.Results)
```

```python
pd.crosstab(df.Results,df.Students)
```

```python
df["Extra Cirricular"]=["Y","Y","N","N","Y"]
df
```

```python
pd.crosstab(df.Students,[df.Results,df["Extra Cirricular"]])
```
