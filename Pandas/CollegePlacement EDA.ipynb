```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
from sklearn.preprocessing import LabelEncoder
```

```python
warnings.simplefilter("ignore")
```

```python
df=pd.read_csv("CollegePlacement.csv")
df.info()
```

```python
df.describe()
```

```python
df.groupby("Gender")["PlacedOrNot"].mean()*100
```

```python
placed_dist=pd.pivot_table(df[["Gender","PlacedOrNot","Stream"]],index=["Gender"],values="PlacedOrNot",aggfunc=np.mean,columns=["Stream"])
placed_dist
```

```python
y=placed_dist.T.reset_index()
```

```python
plt.figure(figsize=(16,3))
plt.subplot(121)
for x,label in zip(["Female","Male"],["Female","Male"]):
    plt.plot(y["Stream"],y[x],label=label)
plt.legend()
plt.xticks(rotation="vertical")
plt.subplot(122)
plt.bar(y["Stream"],y["Female"],color="turquoise",alpha=0.8,label="Female")
plt.bar(y["Stream"],y["Male"],color="royalblue",alpha=0.4,label="Male")
plt.xticks(rotation="vertical")
plt.legend()
plt.show()
```
