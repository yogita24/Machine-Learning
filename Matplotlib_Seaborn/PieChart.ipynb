```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("pie_chart.csv")
df
```

```python
plt.figure(figsize=(6,9))
plt.pie(df["Number of Cars"],labels=df["Engine Fuel Type"],autopct="%.2f%%",textprops={"size":"x-large","fontweight":"bold",'rotation':30,'color':'white'})
plt.legend(loc='best')
plt.title("Pie Chart:Fuel Distribution",fontweight="bold",fontsize=13)
```
